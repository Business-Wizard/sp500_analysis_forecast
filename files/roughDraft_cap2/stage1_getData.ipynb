```python
def grab_tickers():
    
    lst = [
        ['^GSPC', '^IXIC', '^DJI'],
        tickers_sp500(),
        tickers_dow(),
        tickers_nasdaq(),
        tickers_other()
    ]

    for n in range(len(lst)):
        cleans = lst[n]
        clean = [clean.replace('.', '-') for clean in cleans]
        clean = [clean.replace('$', '-') for clean in clean]
        lst[n] = clean
    
    lst_names = [
        'index_tic_lst', 'sp505_tic_lst', 'dow_tic_lst', 'nasdaq_tic_lst', 'other_tic_lst']

    return lst[0], lst[1], lst[2], lst[3], lst[4], lst, lst_names
```

```python
index_tic_lst, sp505_tic_lst, dow_tic_lst, nasdaq_tic_lst, other_tic_lst, all_tic_lst, tic_name_lst = grab_tickers()
```

```python
def storeSaveReview_tic_lst():
    
    tick_lst_count = []

    for i in range(len(all_tic_lst)):
        df = pd.DataFrame(all_tic_lst[i], columns = ['tickers'])
        df.to_csv('data/tic_lsts/' + tic_name_lst[i] + '.csv')

        tick_lst_count.append(len(all_tic_lst[i]))

    print('\n* * * * * * * * REPORT * * * * * * * * * * *')
    for n in range(len(tic_name_lst)):
        print(f'   >{tic_name_lst[n]} Total Tickers = {tick_lst_count[n]}')

    print('\n* All Ticker Lists Have Been Saved To CSV *\n')
```

```python
storeSaveReview_tic_lst()
```

```python
def import_ticker_lsts(run = False):

    if run == False:
        print('NO ACTION TAKEN')

    if run == True:
        ticker_lsts = {}
        all_tic_lst_names = ['index_tic_lst', 'sp505_tic_lst', 'dow_tic_lst', 'nasdaq_tic_lst', 'other_tic_lst']
        
        for n in range(5):
            csv_import = pd.read_csv('data/tic_lsts/' + all_tic_lst_names[n] + '.csv')
            csv_import = csv_import.drop('Unnamed: 0', axis = 1)
            ticker_lsts[n] = (list(csv_import['tickers']))

        print('All Ticker Lists Imported')

    for l in range(len(ticker_lsts)):
        print(len(ticker_lsts[l]))

    return ticker_lsts[0], ticker_lsts[1], ticker_lsts[2], ticker_lsts[3], ticker_lsts[4], ticker_lsts
```

```python
import pandas as pd

from tqdm import tqdm_notebook
from datetime import *

import yfinance as yf
from yahoo_fin.stock_info import get_analysts_info
from yahoo_fin.stock_info import * 

%load_ext autotime


index_lst, sp505_lst, dow_lst, nasdaq_lst, other_lst, ticker_lsts = import_ticker_lsts(True)
```

```python
del other_lst[-1]
clean_other_lst = []

for o in other_lst:
    cleans = list(o)
    string = ''
    for clean in cleans:
        if clean != '-':
            string += clean
        if clean == '-':
            break
    clean_other_lst.append(string)
```

```python
def import_hist(stock_list, save_csv_name, periods, interval):
        for p in periods:
            df = yf.download(stock_list, period = p, interval = interval)['Adj Close']
            df.to_csv('data/hist/' + save_csv_name + p + '_' + interval + '.csv')
```

```python
stock_ticker_lists = [index_lst, dow_lst, sp505_lst, nasdaq_lst, clean_other_lst]
hist_csv_name = ['index_hist_lst_', 'dow_hist_lst_', 'sp505_hist_lst_', 'nasdaq_hist_lst_', 'other_hist_lst_']
interval = '1mo' # 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo
periods = ['ytd', '1y', '2y', '5y', '10y'] 


for n in range(len(stock_ticker_lists)):
    import_hist(stock_ticker_lists[n], hist_csv_name[n], periods, interval)
```

```python
c = 0.0
cMo = 0.0
tally = 0.0

periods = [252, 252 * 2, 252 * 5, 252 * 10, 246]
periods_mo = [11, 12, 12*2, 12*5, 12*10]

for p in periods:
    c = c + p
for p in periods_mo:
    cMo = cMo + p

c = format(c, ",")
cMo = format(cMo, ",")

print(f'\ntotal data points per ticker (DAILY) = {c} \n')
print(f'\ntotal data points per ticker (MONTHLY) = {cMo} \n')

tic_lst_len = [len(index_lst), len(dow_lst), len(sp505_lst), len(nasdaq_lst), len(clean_other_lst)]
tic_lst_name = ['index', 'dow','sp500','nasdaq','other']

for tll in range(len(tic_lst_len)):
    
    new = int(tic_lst_len[tll]) * c
    new_mo = int(tic_lst_len[tll]) * cMo

    print(f'{tic_lst_name[tll]} = {tic_lst_len[tll]} tickers.\n   >(DAILY) data points = {format(new, ",")}\n   >Monthly data points = {format(new_mo, ",")}')

    tally = tally + new
    tally_mo = tally_mo + new_mo

tally = format(tally, ",")
tally_mo = format(tally_mo, ",")

print(f'\ntotal DAILY data pts = {tally} \ntotal MONTHLY data pts = {tally_mo}')
```
